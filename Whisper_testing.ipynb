```python
# Step 3: Prepare folders
import os
os.makedirs("/content/data", exist_ok=True)

# Step 4: Upload your audio file (if needed)
from google.colab import files
uploaded = files.upload()
```

```python
FILENAME = list(uploaded.keys())[0] # convert dict_keys to a list to access elements
print(f"Uploaded file: {FILENAME}")
# After uploading, move it to /content/data if needed
# Example: if your file is "sample.mp3"
!mv {FILENAME} /content/data/{FILENAME}
if os.path.exists(f"/content/data/{FILENAME}"):
  print(f"Success: Found the {FILENAME} file.")
else:
  print(f"Error: The {FILENAME} file isn't found.")
```

```python
# Step 5: Download 'transcribe_diarize.py' from GitHub
# (or manually upload it)
!wget https://raw.githubusercontent.com/DonRichards/whisper_testing/main/transcribe_diarize.py -O /content/transcribe_diarize.py

if os.path.exists("/content/transcribe_diarize.py"):
  print("Success: The transcribe_diarize.py file downloaded.")
else:
  print("Error: The transcribe_diarize.py file didn't download.")
```

```python
# Step 6: Run transcription
# (Adjust "yourfile.mp3" to your uploaded file name)

!python transcribe_diarize.py /content/data/{FILENAME} \
  --output /content/data/output.vtt \
  --model small \
  --num-speakers 2

# Step 7: Download results
if os.path.exists("/content/data/output.vtt"):
  files.download("/content/data/output.vtt")
else:
  print("Error: The VTT file was not generated.")
```
